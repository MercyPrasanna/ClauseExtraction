```python
import sys
import os
import pandas as pd
from readApi import readApi
from utils import json2df, header, footer
```

```python
# configuration parameters

TOP_LINES=5 # number of lines to look for header and footer info on the page
kgram=4     # count of character to consider for k-grams
src_dir = "../dataset/pdfs/"
dst_dir = "../dataset/txt/"
```

```python
ocr_res = "../dataset/ocr_df/"

fls = os.listdir(src_dir)

header_df = pd.DataFrame()
footer_df = pd.DataFrame()

for fl in fls:

    print("working on file:",fl)
    
    #read data using OCR
    dat = readApi(src_dir + fl)
    
    #identify headers and footers
    df = json2df(dat)
    #df.to_csv(ocr_res+fl+".csv") #for debugging only
    
    df = header(df, kgram , TOP_LINES)
    df = footer(df, kgram, TOP_LINES)
    
    #remove headers & footers and save data to text file
    heads = df[df.isHeader]
    feet = df[df.isFooter]
    doc = df[(df.isHeader == False) & (df.isFooter == False)]
    doc_text = doc['text'].to_list()
    with open(dst_dir + fl +'.txt','wt',encoding='utf-8') as f:
        f.write('\n'.join(doc_text))


    header_df = header_df.append(heads)
    footer_df = footer_df.append(feet)
    
header_df.to_csv("header-data.csv")
footer_df.to_csv("footer-data.csv")
```
